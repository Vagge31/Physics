```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x,σ):
    return (x/σ**2)*np.exp(-x**2/(2*σ**2))
```

```python
N=10000
σ=2
def U():
    return np.random.rand()
def T(x):
    return np.sqrt(-2*σ**2 * np.log(1-x))
```

```python
u = np.array([U() for _ in range(N)])
X = T(u)

#Εναλλακτικά με for loop 
#xs = []
#for i in range(N):
    #u = U()
    #x = f(u,2)
    #xs += [x]
    
#Enallaktikos tropos na ypologisoume to pososto kymatwn me ipsos 3<x<5
#n = 0
#for i in range(len(X)): 
#   if 3<X[i] and X[i] <5: 
#       n+=1
#print(n/N)

plt.hist(X,ec='black',bins=20)
plt.savefig("hist.png")
```

```python
#Εναλλακτικα μπορούμε να βρουμε το μεγιστο με αρκετα καλη προσέγγιση με την μεθοδο np.max
#xs = np.linspace(0,10,N)
#ys = f(xs,σ)
#fmax2 = np.max(ys)

fmax=f(2,2) #Το x=2 βρεθηκε λυνοντας την εξισωση f'(x) = 0

hits=[]
for i in range(N):
    x = 10*U()
    y = fmax*U()
    if y <= f(x,σ):
        hits += [x]
plt.hist(hits, ec = 'black',bins=20)
```

```python
#Αθροιστικη συναρτηση κατανομης
def F(x,σ):
    return 1 - np.exp(-x**2/(2*σ**2))
```
